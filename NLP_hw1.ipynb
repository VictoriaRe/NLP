```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
import nltk
nltk.download('punkt')
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train('\n'.join(data))
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
gold = tokenizer.tokenize(data_short)
```

```python
gold
```

```python
len(gold)
```

```python
my_split = sent_tokenize(data_short)
```

```python
my_split
```

```python
print('Gold length - ', len(gold))
print('My split - ', len(my_split))
```

```python
len(set(gold) & set(my_split)) / len(set(gold) | set(my_split)) #метрика для того, чтобы проверить, насколько курто работет модель
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F-measure - ', f1)
```
